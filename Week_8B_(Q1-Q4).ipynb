```python
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor,RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
```

```python
df=pd.read_csv('https://drive.google.com/uc?export=download&id=1tKqddJuIiRqrm7ml2qyVGJlAW3JQjaZX')
```

```python
df=df.dropna()
df=df.drop('Invoice',axis=1)
```

```python
l=[]
for x in df['MSRP']:
    x=x.strip('$')
    temp=x.split(',')
    t=temp[0]+temp[1]
    i=int(t)
    l.append(i)
```

```python
df['MSRP']=l
```

### Use Pandas.get_dummies() on the categorical features : 'Make','Model','Type','Origin', 'DriveTrain'

```python
temp=['Make','Model','Type','Origin', 'DriveTrain']
df=pd.get_dummies(df, prefix=temp)
df
```

```python
features=df.columns[1:]
target=df.columns[0]
X=df[features]
y=df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, 
    test_size=0.30,
    random_state=1,
    shuffle=True
)
```

```python
X.shape
```

```python
y.shape
```

```python
classifiers=[
    BaggingRegressor(random_state=1),
        RandomForestRegressor(random_state=1),
            GradientBoostingRegressor(random_state=1),
                AdaBoostRegressor(random_state=1)
  ]
```

```python
for clf in classifiers:
    clf.fit(X_train,y_train)
    name = clf.__class__.__name__
    print("#"*30)
    print(name)
    print('****Results****')
    print('Emperical/Training Error',clf.score(X_train,y_train))
    print('Genrealisation/Test Error',clf.score(X_test,y_test))
```
