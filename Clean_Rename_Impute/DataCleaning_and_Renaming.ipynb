```python
import pandas as pd 
import numpy as np
from sklearn.impute import SimpleImputer
```

# Read CSV 

```python
df=pd.read_csv('train_data.csv')
```

# Simple Impute 

```python
df.isna().sum()
```

```python
si=SimpleImputer(strategy='most_frequent')
temp=si.fit_transform(df)
df=pd.DataFrame(data=temp,columns=df.columns)
df=df.drop('car',axis=1)
```

# Date Cleaning Function

```python
def clean_date(date):
    i=0
    for x in date:
        # print(x)
        if x=='2days':
            date[i]=48
        else:
            date[i]=10
        i+=1
    return date
```

# Income Cleaning Function

```python
def clean_income(data):    
    for i in range(len(data)):
        x=data[i]
        if x=='₹100000 or More':
            data[i]=100000
        elif x=='Less than ₹12500':
            data[i]=12500
        else:
            s=x.strip().split('-')
            a=float(s[0][1:])
            b=float(s[1][2:])
            data[i]=float((a+b)*0.5) # average calculation
    return data
```

# Age Cleaning Function

```python
def clean_age(age):
    i=0
    for x in age:
        if x=="50plus":
            age[i]=52
        if x=="below21":
            age[i]=18
        i+=1
    return age
```

```python
def clean_misc(data):
    i=0
    for x in data:
        if x=="never":
            data[i]=0
        if x=="less1":
            data[i]=1
        if x=="1~3":
            data[i]=2
        if x=="4~8":
            data[i]=6
        if x=="gt8":
            data[i]=10
        i+=1
    return data
```

# Calling Cleaning Function

```python
temp=df['offer expiration'].copy()
temp=clean_date(temp)
df['offer expiration']=temp
# temp
```

```python
temp=df['income_range'].copy()
temp=clean_income(temp)
df['income_range']=temp
# temp
```

```python
temp=df['age'].copy()
temp=clean_age(temp)
df['age']=temp
# temp
```

```python
cols=['no_visited_Cold drinks','travelled_more_than_15mins_for_offer','Restaur_spend_less_than20',
     'travelled_more_than_25mins_for_offer','travelled_more_than_5mins_for_offer','no_visited_bars',
     'no_Take-aways','Restaur_spend_greater_than20'
     ]
```

```python
for i in cols:
    temp=df[i].copy()
    temp=clean_misc(temp)
    df[i]=temp
```
