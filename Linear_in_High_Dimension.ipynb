# Import Libraries

```python
import seaborn as sns
import numpy as np
```

# Intialize List

```python
x=list(range(-100,101))
y=[]
```

# Creating above y 

```python
for i in x:
    t=((100**2)-(i**2))**0.5
    y.append(t)
    # t=t*(-1)
    # y.append(t)
```

# Saving a copy

```python
# X axis
fun=x.copy()
# Y Axis
tun=y.copy()
```

# Generating Below y

```python
temp=x.copy()
for i in temp:
    x.append(i)
```

```python
for i in temp:
    t=((100**2)-(i**2))**0.5
    t=t*(-1)
    y.append(t)
```

# Transform to Higher Dimension

```python
mon=[x**2 for x in fun] # x axis
pon=[x**2 for x in tun] # y axis
```

# Plot showing implementation

## Show in Lower Dimension 

```python
sns.scatterplot(x=x,y=y)
```

## Changes to Linear in Higher Dimension

```python
sns.scatterplot(x=mon,y=pon)
```
