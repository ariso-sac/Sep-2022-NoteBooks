# Q1 , Descreptive analysis

```python
build_type=df['building_id'].unique()
meter_type=df['meter'].unique()
```

## By Each meter type

```python
d2={}
```

```python
for x in meter_type:
    temp=df[df['meter']==x]
    des=temp['meter_reading'].describe()
    d2[x]=des
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][0]) for k in keys]
plt.title('Count of Readings per type')
plt.xlabel('Meter types')
plt.ylabel('Count')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][1]) for k in keys]
plt.title('Mean of Readings per type')
plt.xlabel('Meter types')
plt.ylabel('Mean')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][2]) for k in keys]
plt.title('STD of Readings per type')
plt.xlabel('Meter types')
plt.ylabel('Standard Deviation')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][5]) for k in keys]
plt.title('Median of Readings per type')
plt.xlabel('Meter types')
plt.ylabel('Median')
sns.barplot(x=keys, y=vals)
```
