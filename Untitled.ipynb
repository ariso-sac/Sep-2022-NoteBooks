# Q1 , Descreptive analysis

```python
build_type=df['building_id'].unique()
meter_type=df['meter'].unique()
```

## By Each meter type

```python
d2={}
```

```python
for x in meter_type:
    temp=df[df['meter']==x]
    des=temp['meter_reading'].describe()
    d2[x]=des
```

### Descriptive Analysis by meter type begins

```python
for x in d2:
    print(x)
    print(d2[x])
    print('block ends')
```

### Visual Analysis by meter type begins

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][0]) for k in keys]
plt.title('Count of Readings per Meter type')
plt.xlabel('Meter types')
plt.ylabel('Count')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][1]) for k in keys]
plt.title('Mean of Readings per Meter type')
plt.xlabel('Meter types')
plt.ylabel('Mean')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][2]) for k in keys]
plt.title('STD of Readings per Meter type')
plt.xlabel('Meter types')
plt.ylabel('Standard Deviation')
sns.barplot(x=keys, y=vals)
```

```python
keys = list(d2.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d2[k][5]) for k in keys]
plt.title('Median of Readings per Meter type')
plt.xlabel('Meter types')
plt.ylabel('Median')
sns.barplot(x=keys, y=vals)
```

# Outlier analysis by meter type

```python
labels, data = d2.keys(), d2.values()
labels=list(labels)
data=list(data)
labels=labels[:-1]
data=data[:-1]
plt.boxplot(data)
plt.ylim(-1000,35000)
plt.xticks(range(1, len(labels) + 1), labels)
plt.show()
```

```python
labels, data = d2.keys(), d2.values()
labels=list(labels)
data=list(data)
labels=labels[-1]
data=data[-1]
plt.boxplot(data)
plt.show()
```

# Hourly Analysis

```python
df['time'].value_counts()
```

```python
unique_hours=df['time'].unique()
```

```python
d3={}
```

```python
for x in unique_hours:
    temp=df[df['time']==x]
    des=temp['meter_reading'].describe()
    d3[x]=des
```

```python
keys = list(d3.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(d3[k][0]) for k in keys]
plt.figure(figsize=(50,15))
plt.title('Count of Readings per Hour')
plt.xlabel('Hours')
plt.ylabel('Count')
sns.barplot(x=keys, y=vals)
```
